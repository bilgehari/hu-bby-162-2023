```python
from google.colab import drive
drive.mount('/content/drive')

dosyaYolu = "/content/drive/MyDrive/Colab Notebooks/veri.txt"

def show_menu():
    print("Katalog Uygulaması\n")
    print("1. Katalogdaki Eserleri Listele")
    print("2. Yeni Eser Ekle")
    print("3. Eser Sil")
    print("4. Eser Güncelle")
    print("5. Katalogu Temizle")
    print("6. Eser Ara")
    print("0. Çıkış")
    print()

def list_books():
    with open(dosyaYolu, "r") as file:
        books = file.readlines()
        if not books:
            print("Katalog boş.")
        else:
            for book in books:
                print(book.strip())

def add_book():
    name = input("Eser Adı: ")
    author = input("Yazar Adı: ")
    year = input("Basım Yılı: ")
    book = f"{name},{author},{year}\n"
    with open(dosyaYolu, "a") as file:
        file.write(book)
    print("Eser başarıyla eklendi.")

def delete_book():
    name = input("Silinecek Eserin Adı: ")
    books = []
    with open(dosyaYolu, "r") as file:
        books = file.readlines()
    with open(dosyaYolu, "w") as file:
        deleted = False
        for book in books:
            if book.split(",")[0] != name:
                file.write(book)
            else:
                deleted = True
        if deleted:
            print("Eser başarıyla silindi.")
        else:
            print("Eser bulunamadı.")

def update_book():
    name = input("Güncellenecek Eserin Adı: ")
    books = []
    with open(dosyaYolu, "r") as file:
        books = file.readlines()
    with open(dosyaYolu, "w") as file:
        updated = False
        for book in books:
            if book.split(",")[0] != name:
                file.write(book)
            else:
                new_name = input("Yeni Eser Adı: ")
                author = input("Yazar Adı: ")
                year = input("Basım Yılı: ")
                updated_book = f"{new_name},{author},{year}\n"
                file.write(updated_book)
                updated = True
        if updated:
            print("Eser başarıyla güncellendi.")
        else:
            print("Eser bulunamadı.")

def clear_catalog():
    confirm = input("Katalog tamamen temizlensin mi? (E/H): ")
    if confirm.lower() == "e":
        with open(dosyaYolu, "w") as file:
            print("Katalog başarıyla temizlendi.")
    else:
        print("Katalog temizlenmedi.")

def search_books():
    search_term = input("Aranacak Eser Adı veya Yazar Adı: ")
    with open(dosyaYolu, "r") as file:
        books = file.readlines()
        found_books = []
        for book in books:
            if search_term.lower() in book.lower():
                found_books.append(book.strip())
        if found_books:
            print("Eşleşen kayıtlar:")
            for found_book in found_books:
                print(found_book)
        else:
            print("Eşleşen kayıt bulunamadı.")

def main():
    while True:
        show_menu()
        choice = input("Bir seçenek girin (0-6): ")
        print()
        if choice == "0":
            print("Katalog Uygulaması sonlandırılıyor...")
            break
        elif choice == "1":
            list_books()
        elif choice == "2":
            add_book()
        elif choice == "3":
            delete_book()
        elif choice == "4":
            update_book()
        elif choice == "5":
            clear_catalog()
        elif choice == "6":
            search_books()
        else:
            print("Geçersiz seçenek. Tekrar deneyin.\n")

if __name__ == "__main__":
    print("Katalog Uygulaması başlatılıyor...\n")
    main()
```
